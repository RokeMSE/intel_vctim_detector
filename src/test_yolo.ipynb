```python
""" !pip install ultralytics """
from ultralytics import YOLO

# Load a model
model = YOLO('yolo11s.pt') 

# Train the model
# data.yml should point to dataset of grid images
results = model.train(
    data='data.yml', 
    epochs=100, 
    imgsz=640, 
    batch=32,
    name='vctim_detector'
)
```

```python
import cv2
from ultralytics import YOLO
import os
import torch

if torch.cuda.is_available():
    device = 'cuda'
    print("CUDA available")
else:
    device = 'cpu'

model = YOLO('../runs/detect/vctim_detector/weights/best.pt')
img_path = './data/test/'
results = model(img_path)

model.export(format='openvino')
for result in results:
    boxes = result.boxes
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0]
        conf = box.conf[0]
        cls = int(box.cls[0])
        label = model.names[cls] 
        
        print(f"{result.path.split('/')[-1]}: Detected {label} with confidence {conf:.2f}")

    # Visualize
    res_plotted = result.plot()
    cv2.imshow("Result", res_plotted)
    cv2.imwrite(os.path.join('../runs/detect/vctim_detector/', result.path.split('/')[-1]), res_plotted)
    cv2.waitKey(0)
```
